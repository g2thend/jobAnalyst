```python
salary = df['salary']
salary
```

```python
sal = salary[ salary != "薪资面议"]
sa = sal.value_counts()
sa.plot.barh(title="薪资发布区间分布", figsize=(10, 10)) 
```

```python
def cut_word(word,method):
    position=word.find('-')
    length=len(word)
    if position !=-1:
        #如果薪资数据内存在-为真,执行本语句,获取-和k之前的数据和-及k之间的数据
        bottomSalary=word[:position-1]
        topSalary=word[position+1:length-1]
    else:
        #如果没有-, 将该数据全部转化为大写后截取数据
        bottomSalary=word[:]
        topSalary=bottomSalary
    if method =='bottom':
        return bottomSalary
    else:
        return topSalary
# 添加新列,获取薪资的最低和最高
df['topSalary']=salary.apply(cut_word,method='top')
df['bottomSalary']=salary.apply(cut_word,method='bottom')

# 函数作用:
# 将薪资示例: 20k-30k分割成20  30 样子,便于分析平均工资
```

```python
df['topSalary'] = df['topSalary'].replace("薪资面议", 0)
df['bottomSalary'] = df['bottomSalary'].replace('薪资面议', 0)
# 将薪资发布中'薪资面议'的条目更换,便于数据处理
```

```python
df['topSalary'] = df['topSalary'].astype('int')
df['bottomSalary'] = df['bottomSalary'].astype('int')
# 将薪资的数据类型转换为可计算类型
```

```python
df['average'] = (df['topSalary'] + df['bottomSalary']) / 2
```

```python
aver = df['average'].value_counts()
```

```python
aver.plot.barh(title="平均薪资", figsize=(8, 8)) 
```

```python
import jieba   
import jieba.posseg as pseg  
from collections import Counter
import os, codecs


##使用字典,内容为一些自定义的有意义的词
#五险一金  年底双薪  绩效奖金  带薪年假  补充医疗保险  等
jieba.load_userdict("dict.txt")
with codecs.open('jobwelfare.csv', 'r', 'utf8') as f:
        txt = f.read()
# words = jieba.cut(txt) 

seg_list = jieba.cut(txt)
c = Counter()
for x in seg_list:
    if len(x)>1 and x != '\r\n':
        c[x] += 1
print('常用词频度统计结果')
for (k,v) in c.most_common():
    print(k, v)
#     print('%s%s %s  %d' % ('  '*(20-len(k)), k, '*'*int(v/3), v))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
wel = pd.DataFrame(c.most_common(), columns=("福利", "数量"))
```

```python
wel1 = wel.set_index("福利")
wel1.head()
```

```python
plt.rcParams['font.family'] = ['simhei']
plt.rcParams['font.size']=10
plt.rcParams['axes.unicode_minus'] = False
wel1.plot.barh(title="热门福利", figsize=(20, 20))
```
